```python
#y is the dependent variable.
y = boston_df['PRICE']

X = boston_df.iloc[:,0:13]

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)


X_train = torch.FloatTensor(X_train.values)
y_train = torch.FloatTensor(y_train.values).unsqueeze(-1)
X_test = torch.FloatTensor(X_test.values)
y_test = torch.FloatTensor(y_test.values).unsqueeze(-1)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)



```

```python
#Create the model


device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") #cpu or gpu :)

# Assume that we are on a CUDA machine, then this should print a CUDA device:
print("Executing the model on :",device)

# Neural net
net = torch.nn.Sequential(
        torch.nn.Linear( X_train.shape[1], 200),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(200, 100),
        torch.nn.LeakyReLU(),
        torch.nn.Linear(100, 1),
    )

#Optimizer
optimizer = torch.optim.Adam(net.parameters(), lr=0.01) 

#Loss function
loss_func = torch.nn.MSELoss()  # Regression mean squared loss

BATCH_SIZE = 202
EPOCH = 750 #how many times we go through sample

torch_dataset = torch.utils.data.TensorDataset(X_train, y_train) #create dataset 


loader = torch.utils.data.DataLoader(
    dataset=torch_dataset, 
    batch_size=BATCH_SIZE, 
    shuffle=True, num_workers=2,)
```

```python
#Training 
running_loss=0


# start training
for epoch in range(EPOCH):
    for step, (bx, by) in enumerate(loader): # for each training step
        prediction = net(bx)     # Input x and predict based on x
    
        loss = loss_func(prediction, by) #Calculate loss given nnoutput, target
        
        optimizer.zero_grad()   # clear gradients for next train (ALWAYS!)
        loss.backward()         # backpropagation, compute gradients <3
        optimizer.step()        # apply gradients
        running_loss += loss.item() 
    print('Epoch {}'.format(epoch+1), "loss: ",running_loss)
    running_loss = 0.0
        
```

```python

result = net(X_train)
pred=result.data[:,0].numpy()

sklearn.metrics.r2_score(pred,y_train)
```

```python
result = net(X_test)
pred=result.data[:,0].numpy()

sklearn.metrics.r2_score(pred,y_test)
```
